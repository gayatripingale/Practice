```python
import requests as req
import json
```

```python
url = "http://nyt-mongo-scraper.herokuapp.com/api/headlines"
response = req.get(url)
```

```python
# Print the response object to the console
print(response.status_code)
```

```python
j = response.json()
len(j)
```

```python
#response = req.get(url).json()
#Because the data is too big,the jupyter notebook won't print the following print command, and, will give some error. 
#So, just pring one by using [0]

print(json.dumps(response.json()[0], indent=4, sort_keys=True))
```

```python
response_json = response.json()
response_json
```

```python
# Print first and last articles
print("The first response is " + json.dumps(response_json[0], indent=2) + ".")
print("The last response is " + json.dumps(response_json[-1], indent=2) + ".")
```
